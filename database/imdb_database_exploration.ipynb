```python
import pandas as pd
import os
```

```python
data_folder = 'imdb_data'
data_files = []
for _, _, files in os.walk(data_folder, topdown=False):
   for name in files:
      data_files.append(os.path.join(data_folder, name))
```

```python
all_info_merge_1 = pd.merge(all_info,titles_basic_pf,how='left',on='tconst')

all_info_merge_1.head()
```

```python
# Drop last two columns
all_info_merge_2 = all_info_merge_1.drop(['tableName','tableDescription'],axis=1)
```

```python
all_info_merge_2.head()
```

```python
# get ratings of each episode

all_info_merge_3 = pd.merge(all_info_merge_2,titles_ratings_pf,how='left',on='tconst')

all_info_merge_3.head()
```

```python
all_info_merge_4 = all_info_merge_3.drop(['tableName','tableDescription'],axis=1)

all_info_merge_4.head()
```

```python
# Rename parentTconstant column to prepare for last merge

all_info_merge_5 = all_info_merge_4.rename(index=str,columns={'tconst':'episode_id','parentTconst':'tconst'})

all_info_merge_5.head()
```

```python
# Merge Series name onto table

all_info_merge_6 = pd.merge(all_info_merge_5,titles_basic_pf,how='left',on='tconst')

all_info_merge_6.head()
```

```python
all_info_merge_7 = all_info_merge_6.drop(['tableName','tableDescription'],axis=1)

all_info_merge_7.head()
```

```python
all_info_merge_7.columns
```

```python
# Rename columns

all_info_merge_8 = all_info_merge_7.rename(index=str,columns=
                                           {'tconst':'series_id',
                                            'parentTconst':'tconst',
                                            'titleType_x': 'titleType_episode',
                                            'primaryTitle_x': 'primaryTitle_episode',
                                           'originalTitle_x':'originalTitle_episode',
                                           'isAdult_x':'isAdult_episode',
                                           'startYear_x':'startYear_episode',
                                           'endYear_x':'endYear_episode',
                                           'runtimeMinutes_x': 'runtimeMinutes_episode',
                                           'genres_x': 'genres_episode',
                                           'titleType_y':'titleType_series',
                                           'primaryTitle_y': 'primaryTitle_series',
                                           'originalTitle_y':'originalTitle_series',
                                           'isAdult_y':'isAdult_series',
                                           'startYear_y': 'startYear_series',
                                           'endYear_y':'endYear_series',
                                           'runtimeMinutes_y':'runtimeMinutes_series',
                                           'genres_y':'genres_series'})
                                                                         
all_info_merge_8.head()
```

```python
# Sort dataframe by series_id
```

```python
all_info_merge_9 = all_info_merge_8.sort_values(by=['series_id'])

all_info_merge_9.head()
```

```python
all_info_merge_9.columns
```

```python
# take only columns needed to plot

all_info_merge_10 = all_info_merge_9.filter(
                                    ['series_id','primaryTitle_series','episode_id','primaryTitle_episode','seasonNumber','episodeNumber',
                                    'isAdult_episode','startYear_episode','endYear_episode',
                                    'runtimeMinutes_episode','genres_episode','averageRating','numVotes'], axis = 1)


all_info_merge_10.head()
```

```python
# Sort the series, and season, episode
```
